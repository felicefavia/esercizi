```python
# Esercizi:
# Creare una classe Person , estenderla con due sottoclassi Dipendente e Studente
# Estendere Dipendenti con 2 sottoclassi Docente e Freelance

class Person: 
    # Costruttore per creare un'istanza della classe
    def __init__(self, name, surname, age):
        self.name = name
        self.surname = surname
        self.age = age
    
    def __str__(self):
        return f"{self.name} {self.surname}, {self.age} anni"

class Dipendente(Person):
    def __init__(self, name, surname, age, salary):
        super().__init__(name, surname, age)
        self.salary = salary

    def __str__(self):
        return f"{super().__str__()}, stipendio: {self.salary} euro"

class Studente(Person):
    def __init__(self, name, surname, age, university):
        super().__init__(name, surname, age)
        self.university = university

    def __str__(self):
        return f"{super().__str__()}, studente presso: {self.university}"

person = Person("Mario", "Rossi", 30)
dicente = Dipendente("Luigi", "Bianchi", 40, 3000)
studente = Studente("Anna", "Verdi", 20, "Università di Milano")

print(person)
print(dicente)
print(studente)
```

```python
class Docente(Dipendente):
    def __init__(self, name, surname, age, salary, subject):
        super().__init__(name, surname, age, salary)
        self.subject = subject

    def __str__(self):
        return f"{super().__str__()}, insegna: {self.subject}"
    
class Freelance(Dipendente):
    def __init__(self, name, surname, age, salary, project):
        super().__init__(name, surname, age, salary)
        self.project = project

    def __str__(self):
        return f"{super().__str__()}, lavora al progetto: {self.project}"
    
docente = Docente("Giovanni", "Neri", 50, 4000, "Matematica")
freelance = Freelance("Sara", "Gialli", 35, 3500, "Sviluppo software")

print(docente)
print(freelance)
```
