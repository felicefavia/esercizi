```python
# Esercizio 1: Creare una lista di numeri e calcolarne la somma

numbers = [10 ,22, 54, 3, 18, 12, 1]

# prima soluzione con for

total_sum = 0
for num in numbers:
    total_sum += num

print("( for ) La somma dei numeri è:", total_sum)

#  seconda soluzione

total_sum = 0
total_sum = sum(numbers)

print("( sum ) La somma dei numeri è:", total_sum)
```

```python
# Installate NumPy e create un array. Stampate la somma di tutti gli elementi
import numpy as np

array = np.array([1, 2, 3, 4, 5])

total_sum = np.sum(array)

print("La somma degli elementi dell'array è:", total_sum)
```

```python
# Esercizio 3: Verifica del numero più grande
# Scrivete un programma che prenda due numeri, confronti quale sia maggiore e stampi un messaggio appropriato.

num1 = float(input("Inserisci il primo numero: "))
num2 = float(input("Inserisci il secondo numero: "))

if num1 > num2:
    print(f"{num1} è maggiore di {num2}.")
elif num1 < num2:
    print(f"{num2} è maggiore di {num1}.")
else:
    print("I due numeri sono uguali.")
```

```python
#Esercizio 4: Verifica della maggiore età
#Scrivete un programma che controlli se una persona è maggiorenne (18 anni o più) e stampi un messaggio appropriato.

age = int(input("Inserisci la tua età:"))

if age >= 18:
    print("Sei maggiorenne.")
else:
    print("Sei minorenne.")
```

```python
# Esercizio 5: Parità di un numero
# Scrivete un programma che prenda un numero e verifichi se è pari o dispari.

num = int(input("Inserisci un numero: "))

if num % 2 == 0:
    print(f"{num} è pari.")
else:
    print(f"{num} è dispari.")
```

```python
# Esercizio 6: Trova il numero più grande tra tre senza usare if
# Scrivete un programma che prenda tre numeri e trovi il più grande tra di loro

num1 = float(input("Inserisci il primo numero: "))
num2 = float(input("Inserisci il secondo numero: "))
num3 = float(input("Inserisci il terzo numero: "))

largest = max(num1, num2, num3)

print(f"Il numero più grande è: {largest}")
```
