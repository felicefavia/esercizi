```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env')

client = OpenAI()

completion = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {"role": "user", "content": "Ciao mi consigli cosa mangiare stasera, qualcosa di leggero."}
    ]
)

print(completion.choices[0].message.content)
```
