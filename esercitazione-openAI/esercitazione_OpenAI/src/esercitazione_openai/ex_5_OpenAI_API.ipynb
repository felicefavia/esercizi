```python

# TODO Inizializza il client di Open AI
import os 
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env')

client = OpenAI()

```

```python
input_text = input("Enter text: ")


complention = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Sei un generatore di domande. Quando ti danno del testo, tu in base al testo fai domande inerenti all'argomento"},
        {"role": "user", "content": input_text}
    ],
)

print(complention.choices[0].message.content)
```
